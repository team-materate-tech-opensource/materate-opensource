```python
import pandas as pd
import numpy as np
```

```python
data = 'algebra_2005_2006_test.txt'
```

```python
data = pd.read_table(data)
```

```python
hierarchy = data['Problem Hierarchy']
units, sections = [], []
for i in range(len(hierarchy)):
    units.append(hierarchy[i].split(',')[0].strip())
    sections.append(hierarchy[i].split(',')[1].strip())

#adding units and sections as columns to the train data
data['Problem Unit'] = pd.Series(units, index=data.index)
data['Problem Section'] = pd.Series(sections, index=data.index)
#rearranging column oreder
cols = data.columns.tolist()
cols = cols[0:3]+cols[-2::]+cols[3:-2]
data = data[cols]
```

```python
data.head()
```

```python
pu = data['Problem Unit']
ps = data['Problem Section']
ps = ps.unique()
```

```python
for i in range(len(ps)):
    ps[i] = ps[i].split(" ")[1]
```

```python
a = []
for i in range(len(ps)):
    c = []
    b = ps[i].split('-')
    if len(b)!=1:
        c.append(b[-2])
        c.append(b[-1])
        a.append(c) 
```

```python
a = np.array(a)
a[:,0]
```

```python
i = 1
lines = []
skill  = "1 1 1 1"
lines.append(str(i))
lines.append("Algebra")
lines.append(skill)
lines.append('-')
lines.append('-')
concept_map_key = {}
concept_map_key["Algebra"] = 1
for concept in a[:,0]:
    i = i+1
    lines.append(str(i))
    lines.append(concept)
    lines.append(skill)
    lines.append('1')
    lines.append('1')
    concept_map_key[concept]=i
```

```python
for j in range(len(a[:,1])):
    i = i+1
    lines.append(str(i))
    lines.append(a[j,1])
    lines.append(skill)
    lines.append(str(concept_map_key[a[j,0]]))
    lines.append('1')
```

```python
for i in range(len(lines)):
    lines[i]=lines[i]+'\n'
```

```python
out = open("concepts_data.txt",'w+')
out.writelines(lines)
out.close()
```
